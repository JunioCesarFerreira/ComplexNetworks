### Introdução


As redes complexas são estruturas que podem ser encontradas em uma variedade de sistemas, desde redes sociais até redes de computadores e redes biológicas. A análise dessas redes tem sido amplamente estudada nos últimos anos. Uma das abordagens mais comuns para entender as redes complexas é estudar suas propriedades topológicas de baixa ordem, como grau médio, coeficiente de agrupamento e distância média entre os vértices. Essas medidas fornecem informações valiosas sobre a estrutura global e local das redes, mas podem não ser suficientes para capturar aspectos mais sutis e complexos da organização da rede. Nesse contexto, surgem as estruturas topológicas de alta ordem, que se referem a padrões mais intrincados e abrangentes presentes nas redes complexas. No âmbito deste trabalho que se dedica ao estudo das estruturas topológicas de alta ordem em redes complexas, destaca-se a importância da investigação dos complexos simpliciais. Um complexo simplicial é um objeto matemático que captura as relações de conectividade entre conjuntos de vértices em uma rede complexa. Ele é composto por simplexos, que são conjuntos de vértices completamente conectados entre si. Essa estrutura hierárquica permite a representação de padrões de alta ordem. Ao estudar os complexos simpliciais em redes complexas, busca-se entender as interações complexas entre múltiplos vértices e como essas interações contribuem para a formação de estruturas topológicas significativas. Essa abordagem permite explorar os padrões emergentes nas redes complexas e sua relação com a funcionalidade e dinâmica do sistema em estudo.


### Simplexos


Matematicamente, um simplexo é definido como o envoltório convexo de um conjunto de pontos geométricamente independentes em um espaço euclidiano. Em outras palavras, é a menor figura geométrica que contém todos os pontos do conjunto e é formada por segmentos de linha, triângulos, tetraedros, etc., dependendo da dimensão do espaço em que está inserido.

Formalmente definimos para $ p\le n $:

$ \Delta^p:=[x_0,...,x_p]:=
        \bigg\{ 
            \sum_{i=0}^{p}{t_ix_i} : {0\leq t_i \leq 1} \textrm{ e } \sum_{i=0}^{p}{t_i=1}
            \bigg\}\subset\mathbb{R}^n $<br>
            
sendo $ \Delta^p $ chamado de p-simplexo.

Observe que: um 0-simplexo é um ponto, um 1-simplexo é um segmento de reta que liga dois pontos, um 2-simplexo é um triângulo cheio, um 3-simplexo é um tetraedro cheio e um n-simplexo pode ser considerado uma generalização de um polígono convexo para dimensões superiores.

Em um simplexo, uma face é um subconjunto dos vértices do simplexo que forma um simplexo de dimensão menor. Em outras palavras, uma face é um simplexo obtido removendo um ou mais vértices do simplexo original. Por exemplo, considere um tetraedro, que é um simplexo de dimensão 3 com quatro vértices. As faces desse tetraedro seriam:

- As quatro faces de dimensão 2 seriam os triângulos formados pelos subconjuntos de três vértices.
- As seis faces de dimensão 1 seriam as arestas formadas pelos subconjuntos de dois vértices.
- As quatro faces de dimensão 0 seriam os vértices individuais.


### Complexos simpliciais


Um complexo simplicial $K$ é um conjunto de simplexos que satisfazem as seguites propriedades:

- Toda face de um simplexo pertencente a $K$ também pertence a $K$.
- Toda interseção não vazia de simplexos $\sigma_1,\sigma_2\in K$ é uma face de ambos simplexos $\sigma_1$ e $\sigma_2$. 

A hierarquia de inclusão é um aspecto fundamental dos complexos simpliciais. Isso significa que se um simplexo está presente no complexo, todos os seus sub-simplexos também devem estar presentes. Por exemplo, se um triângulo está presente, todas as suas arestas e vértices também devem estar presentes.

#### Definição p-esqueleto
Chamamos de p-esqueletos de um complexo simplicial, as subestruturas formadas pelos simplexos de dimensão igual ou menor que p. O p-esqueleto é obtido removendo os simplexos de dimensão maior que p, mantendo apenas os simplexos de dimensão até p. Isso permite estudar as propriedades do complexo simplicial em uma determinada dimensão específica.

#### Definição dimensão simplicial
A dimensão simplicial de um complexo simplicial é o inteiro não negativo máximo d tal que o complexo contenha um d-simplicial não vazio, mas não contenha um (d+1)-simplicial não vazio. 

### Homologia simplicial


A homologia simplicial é um conceito fundamental na área da topologia algébrica que estuda as propriedades de um complexo simplicial por meio de uma sequência de grupos abelianos chamada grupos de homologia. Esses grupos de homologia fornecem informações importantes sobre a estrutura topológica do complexo simplicial.

A ideia básica por trás da homologia simplicial é capturar a presença de "cavidades" no complexo simplicial. Esses cavidades podem ser entendidas como regiões que não podem ser deformadas continuamente para se tornarem simples ou vazias. A homologia simplicial permite quantificar e classificar essas cavidades de diferentes dimensões.

Para calcular a homologia simplicial, construímos uma sequência de grupos abelianos chamados grupos de homologia simplicial. Cada grupo de homologia, denotado como $H_n$, mede a presença de cavidades de dimensão $n$ no complexo simplicial. O grupo $H_0$ mede o número de componentes conectadas, $H_1$ mede o número de buracos circulares, $H_2$ mede o número de cavidades tridimensionais, e assim por diante.

Os grupos de homologia são calculados através da aplicação de dois operadores matemáticos chamados operador bordo e operador ciclo. Esses operadores capturam a relação entre as faces e os limites dos simplexos no complexo simplicial. O cálculo da homologia é feito por meio da construção de uma sequência exata de grupos e mapeamentos entre eles.

A homologia simplicial é uma poderosa ferramenta para a análise e classificação de complexos simpliciais. Ela permite identificar características topológicas importantes, como a presença de buracos, conectividade e estruturas de cavidades em diferentes dimensões. A homologia simplicial é utilizada em TDA (Topological Data Analyses), e tem aplicações em diversas áreas, incluindo processamento de imagens, análise de redes complexas, reconhecimento de padrões e modelagem de dados.


### Números de Betti


Os números de Betti são uma medida numérica importante na teoria da homologia simplicial. Eles fornecem informações sobre o número e a dimensionalidade das cavidades em um complexo simplicial. Eles são úteis para descrever e classificar diferentes tipos de complexos simpliciais, bem como para comparar a topologia de diferentes espaços.

Para entender os números de Betti, consideramos a sequência de grupos de homologia simplicial 

$H_0, H_1, H_2, ..., H_n$. 

O número de Betti de ordem k, denotado como $β_k$, é igual a dimensão do grupo de homologia $H_k$. Em outras palavras, $β_k$ representa o número de cavidades de dimensão $k$ presentes no complexo simplicial.

Os números de Betti são frequentemente usados para descrever as características topológicas de um complexo simplicial. Por exemplo, o número de Betti $β_0$, também conhecido como o número de componentes conectadas, indica quantas partes distintas e desconectadas o complexo simplicial possui. O número de Betti $β_1$ está relacionado ao número de buracos de dimensão 1 presentes no complexo. Para dimensões superiores, os números de Betti $β_k$ representam o número de cavidades de dimensão k.

Uma propriedade importante dos números de Betti é que eles são invariantes topológicos. Isso significa que, mesmo que o complexo simplicial seja sujeito a deformações ou transformações contínuas, os números de Betti permanecerão os mesmos, indicando as características essenciais da sua estrutura topológica.


### Característica de Euler

A característica de Euler é uma medida fundamental na topologia que descreve a estrutura de um objeto topológico, como uma superfície, uma rede ou um complexo simplicial. É uma quantidade que está relacionada ao número de vértices, arestas e faces de um objeto e fornece informações valiosas sobre sua topologia.

A característica de Euler, denotada por $\chi$, é calculada como a diferença entre o número de vértices (V), o número de arestas (E) e o número de faces (F) de um objeto, seguindo a fórmula χ = V - E + F. Em outras palavras, é uma medida que captura a relação entre os componentes fundamentais de um objeto e nos permite entender melhor sua estrutura topológica.

Seja $K$ um complexo simplicial, denotamos $|K_n|$ a cardinalidade do conjunto dos $n-$simplexos em $K$. A característica de Euler de $K$ é pode ser calculada por:

$ \chi(K)=\sum_{n\ge 0}(-1)^n |K_n|. $

A característica de Euler está intimamente ligada aos números de Betti, pois temos a seguinte relação:

$ \chi(K)=\sum_{n\ge 0}(-1)^n \beta_n,  $

sendo $\beta_n$ o n-ésimo número de Betti do complexo simplicial $K$.

# Aplicação em Redes Complexas

Existem duas abordagens distintas para aplicação da homologia simplicial em redes complexas: a extração do complexo simplicial através dos cliques da rede e a utilização de métricas na rede para gerar a homologia persistente.

Na primeira abordagem, a rede complexa é convertida em complexo simplicial, onde cada clique da rede (conjunto completo de vértices mutuamente conectados) é mapeado para um $(n-1)$-simplexo, sendo $n$ a dimensão do clique. A partir disso, a homologia simplicial é aplicada para identificar características topológicas da rede complexa.

Na segunda abordagem, métricas de rede, como coeficiente de agrupamento, centralidade ou caminhos mínimos, são utilizadas para atribuir valores a cada vértice ou aresta na rede. Esses valores métricos são então usados para construir um complexo simplicial ponderado, onde os pesos das arestas correspondem aos valores das métricas. A homologia simplicial é aplicada a esse complexo ponderado para obter informações sobre as características topológicas persistentes na rede ao long de diferentes escalas.

## Homologia Persistente

A homologia persistente é uma extenção da homologia simplicial que permite a análise de características topológicas em várias escalas. Ela captura a presença de características topológicas em diferentes níveis de complexidade e fornece informações sobre a persistência dessas características ao longo de diferentes intervalos. Isso é particularmente útil para revelar padrões de conectividade em redes complexas, como comunidades ou hierarquias.

Existem vários métodos para construção de complexos simpliciais na homologia persistente, apenas a título de exemplo, apresentamos a seguir a definição do complexo de Vietoris-Rips:

### Definição
Seja $X$ um espaço métrico, com métrica $d$. Então o complexo de Vietoris-Rips para $X$, dado um parâmetro $\epsilon$, denotado $VR(X,\epsilon)$, será o complexo simplicial cujos vértices estão em $X$, sendo $\{x_0,x_1,...,x_k\}$ espandem um $k$-simplexo se, somente se, $d(x_i,x_j)\le\epsilon$ para todo $0\le i,j\le k$.

Uma filtração de um complexo simplicial $K$, é uma sequência crescente de subcomplexos iniciando pelo vazio e terminando no complexo $K$, isto é,

$\empty\subset K_0\subset K_1\subset\cdots\subset K_n=K.$

### Função de abertura de arquivos

```python
import networkx as nx

def OpenNetwork(filename, digraph=True): 
    """Abre arquivos de redes complexas baixados de http://konect.cc/networks/

    Args:
        filename (string): Nome do arquivo
        digraph (bool, optional): Indica se retorno será um grafo direcionado ou não. Defaults to True.

    Returns:
        networkx.: networkx Digrafo ou Graph dependendo do parâmetro digraph.
    """
    G = nx.DiGraph()

    with open(filename, 'r') as file:
        for line in file:
            if not line.startswith("%"):
                source, target, capacity = line.split()
                source = int(source)
                target = int(target)
                capacity = int(capacity)
                G.add_edge(source, target, capacity=capacity)
    if digraph:
        return G
    else:
        return G.to_undirected()
```

### Função para cômputo de algumas medidas básicas de redes complexas

```python
import numpy as np
import networkx as nx

def momment_of_degree_distribution(G,m):
    """Calcula momento da distribuição do grau

    Args:
        G (networkx.Graph): Rede complexa
        m (int): momento a ser calculado

    Returns:
        float: n-ésimo momento da distribuição do grau
    """
    M = 0
    N = len(G)
    for i in G.nodes:
        M = M + G.degree(i)**m
    M = M/N
    return M

def network_measures(G):    
    """Calcula algumas medidas de redes complexas

    Args:
        G (networkx.Graph): Rede complexa 

    Returns:
        array: grau médio, 2º momento do grau, variância do grau, clustering médio, menor caminho médio, coef. assortatividade do
    """
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
    G = G.subgraph(Gcc[0])
    
    vk = dict(G.degree())
    vk = list(vk.values())
    vk = np.array(vk)
    k1 = momment_of_degree_distribution(G,1)
    k2 = momment_of_degree_distribution(G,2)
    variance = k2 - k1**2
    av_cl = nx.average_clustering(G)
    l = nx.average_shortest_path_length(G)
    r=nx.degree_assortativity_coefficient(G)

    return  k1, k2, variance, av_cl, l, r
```

### Funções relacionadas a análise de complexos simpliciais

```python
import numpy as np
import gudhi

def get_simplicial_complex(G):
    """Gera estrutura de complexo simplicial para uma rede complexa.

    Args:
        G (networkx.Graph): Rede complexa

    Returns:
        gudhi.SimplexTree: _description_
    """
    complex = gudhi.SimplexTree()
    
    cliques = list(nx.enumerate_all_cliques(G))    
    
    # Cria o complexo simplicial a partir das listas de cliques    
    for clique in cliques: 
        complex.insert(clique)
            
    return complex

def betti_numbers_of_complex(complex, vector_len=0):    
    """Calcula números de Betti de um complexo simplicial

    Args:
        complex (gudhi.SimplexTree): Complexo simplicial
        vector_len (int, optional): Quando diferente de 0 fixa tamanho do vetor de saída. Defaults to 0.

    Returns:
        array: vetor com números de Betti em seus respectivos indíces.
    """
    # Calcula homologia do complexo simplicial
    complex.persistence(homology_coeff_field=2, min_persistence=-1.0, persistence_dim_max=True)

    # Retorna números de Betti da homologia calculada
    betti_num = complex.betti_numbers()
    
    if vector_len !=0:
        # fixa comprimento do vetor a ser retornado
        betti_num = betti_num[:vector_len]
        if (len(betti_num) < vector_len):
            betti_num = np.pad(betti_num, (0, vector_len - len(betti_num)), mode='constant')
        
    return betti_num

def betti_numbers(G, vector_len=0):
    """Calcula números de Betti de uma rede complexa

    Args:
        G (networkx.Graph): Rede complexa.
        vector_len (int, optional): Quando diferente de 0 fixa tamanho do vetor de saída. Defaults to 0.

    Returns:
        array: vetor com números de Betti em seus respectivos indíces.
    """    
    complex = get_simplicial_complex(G) # Complexo simplicial

    return betti_numbers_of_complex(complex, vector_len)

def simplexes_resume(simplex_tree, vector_len=0):
    """Resumo dos simplexos, quantidade em cada dimensão

    Args:
        simplex_tree (gudhi.SimplexTree): Complexo simplicial
        vector_len (int, optional): Quando diferente de 0 fixa tamanho do vetor de saída. Defaults to 0.

    Returns:
        list: vetor com quantidade de simplexos em cada dimensão.
    """
    simplexes_len={}
    for simplex,filtration in simplex_tree.get_filtration():
        if(len(simplex)-1 not in simplexes_len):
            simplexes_len[len(simplex)-1]=1
        else:
            simplexes_len[len(simplex)-1]+=1    

    simplexes_resume=[]
    for item in simplexes_len.items():
        simplexes_resume.append(item[1])
        
    if vector_len != 0:
        simplexes_resume = simplexes_resume[:vector_len]
        if (len(simplexes_resume) < vector_len):
            simplexes_resume = np.pad(simplexes_resume, (0, vector_len - len(simplexes_resume)), mode='constant')
    
    return simplexes_resume
```

### Primeiro exemplo

O exemplo a seguir apresenta duas pequenas redes que, podem parecer semelhantes quando consideramos apenas as medidas usuais. No entanto, a análise de sua topologia revela diferenças distintas. Observe que os números de Betti da rede G1 são iguais a 0 para dimensões maiores ou iguais a 1. Isto ocorre, pois o grafo G1 é contrátil. Enquanto para a rede G2 temos $\beta_1=1$, pois G2 não é contrátil, na verdade G2 é topologicamente equivalente a um círculo.

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

A1 = np.array([
#    0 1 2 3 4 5 6
    [0,1,0,0,0,0,1], # 0
    [1,0,1,1,0,0,0], # 1
    [0,1,0,0,0,0,0], # 2
    [0,1,0,0,1,1,0], # 3
    [0,0,0,1,0,0,0], # 4
    [0,0,0,1,0,0,0], # 5
    [1,0,0,0,0,0,0]  # 6
    ]) 
G1 = nx.from_numpy_array(np.array(A1), create_using=nx.Graph)  

A2 = np.array([
#    0 1 2 3 4 5 6
    [0,1,1,0,0,0,1], # 0
    [1,0,0,1,0,0,0], # 1
    [1,0,0,1,0,0,0], # 2
    [0,1,1,0,1,0,0], # 3
    [0,0,0,1,0,1,0], # 4
    [0,0,0,0,1,0,0], # 5
    [1,0,0,0,0,0,0]  # 6
    ]) 
G2 = nx.from_numpy_array(np.array(A2), create_using=nx.Graph)  

plt.subplot(1,2,1)
plt.title("G1")
nx.draw_kamada_kawai(G1, with_labels=True, edgecolors='black')
plt.subplot(1,2,2)
plt.title("G2")
nx.draw_kamada_kawai(G2, with_labels=True, edgecolors='black')
plt.show()

results_G1 = network_measures(G1)
results_G2 = network_measures(G2)

print(f"network measures G1 = {[round(result, 2) for result in results_G1]}")
print(f"network measures G2 = {[round(result, 2) for result in results_G2]}")

betti_G1 = betti_numbers(G1)
betti_G2 = betti_numbers(G2)

print("betti numbers G1:", betti_G1)
print("betti numbers G2:", betti_G2)
```

Nesta seção, realizamos alguns testes com o objetivo de identificar visualmente diferenças inerentes entre alguns modelos geradores de redes.

```python
import networkx as nx

def test_generate_networks(n_nodes, av_degree, n_networks):
    """Gera redes Erdos-Reni, Barabasi-Albert e Watts-Strogatz para comparar propriedades topológicas de alta ordem.

    Args:
        n_nodes (int): Número de vértices para geração de redes
        av_degree (float): Grau médio para geração de redes
        n_networks (int): Quantidade de redes a serem geradas

    Returns:
        void: sem retorno
    """
    
    # Parâmetros calculados
    p = av_degree/(n_nodes-1)
    m = int(av_degree/2)
    k = int(av_degree)

    def m_component(G):
        Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
        return G.subgraph(Gcc[0])
    
    def draw_network(G):
        pos = nx.kamada_kawai_layout(G)
        d = dict(G.degree())
        nx.draw(G, pos, node_size=[v * 5 for v in d.values()], edgecolors='black', with_labels=False)

    for _ in range(0, n_networks):
        # Erdos-Renyi
        GER = nx.erdos_renyi_graph(n_nodes, p, seed=None, directed=False)
        # Barabasi-Albert
        GBA = nx.barabasi_albert_graph(n_nodes, m, seed=None)
        # Small-world
        GWS = nx.watts_strogatz_graph(n_nodes, k, p, seed=None)
    
        plt.figure(figsize=(12, 4))

        plt.subplot(1,3,1)
        plt.title("GER")
        draw_network(GER)
        
        plt.subplot(1,3,2)
        plt.title("GBA")
        draw_network(GBA)
            
        plt.subplot(1,3,3)
        plt.title("GWS")
        draw_network(GWS)
        
        plt.show()
        
        complex_GER = get_simplicial_complex(GER)
        complex_GBA = get_simplicial_complex(GBA)
        complex_GWS = get_simplicial_complex(GWS)

        print("Betti numbers(GER) = ", betti_numbers_of_complex(complex_GER))
        print("Betti numbers(GBA) = ", betti_numbers_of_complex(complex_GBA))
        print("Betti numbers(GWS) = ", betti_numbers_of_complex(complex_GWS))
        
        print ("Simplexes resume(GER):", simplexes_resume(complex_GER))
        print ("Simplexes resume(GBA):", simplexes_resume(complex_GBA))
        print ("Simplexes resume(GWS):", simplexes_resume(complex_GWS))
            
        GER = m_component(GER)
        GBA = m_component(GBA)
        GWS = m_component(GWS)

        er_measures = network_measures(GER)
        ba_measures = network_measures(GBA)
        ws_measures = network_measures(GWS)

        print(f"Network measures(GER) = {[round(result, 2) for result in er_measures]}")
        print(f"Network measures(GBA) = {[round(result, 2) for result in ba_measures]}")
        print(f"Network measures(GWS) = {[round(result, 2) for result in ws_measures]}")
```

### Função para gerar redes e comparar comportamento dos números de Betti

A função a seguir gera redes Erdos-Reni, Barabasi-Albert e Watts-Strogatz. Fixando a quantidade de vértices, variamos o grau médio dentro de um intervalo, plotando ao final um comparativo da evolução de cada número de vértices conforme o grau médio aumenta.

Alerta! Dependendo dos parâmetros de entrada essa função pode levar algumas horas para ser executada.

```python
import networkx as nx
import matplotlib.pyplot as plt

def test_generate_networks_plot_degree_betti_num(n_nodes, av_degree_start, av_degree_stop, n_networks, max_dim):
    """Gera redes Erdos-Renyi, Barabasi-Albert e Watts-Strogatz para comparar propriedades topológicas de alta ordem.

    Args:
        n_nodes (int): Número de vértices.
        av_degree_start (int): Grau médio inicial.
        av_degree_stop (int): Grau médio final.
        n_networks (int): Número de redes a serem geradas.
        max_dim (int): Dimensão máxima a ser plotada.
    """
    vector_len = max_dim+1
    X = []
    Y_GER = []
    Y_GBA = []
    Y_GWS = []
    for av_degree in range(av_degree_start, av_degree_stop):    
        X.append(av_degree)
        betti_GER = np.zeros(vector_len)
        betti_GBA = np.zeros(vector_len)
        betti_GWS = np.zeros(vector_len)
        for _ in range(0, n_networks):
            p = av_degree/(n_nodes-1)
            m = int(av_degree/2)
            k = int(av_degree)
            # Erdos-Reni
            GER = nx.erdos_renyi_graph(n_nodes, p, seed=None, directed=False)
            # Barabasi-Albert
            GBA = nx.barabasi_albert_graph(n_nodes, m, seed=None)
            # Small-world
            GWS = nx.watts_strogatz_graph(n_nodes, k, p, seed=None)
            
            betti_GER = np.array(betti_GER) + np.array(betti_numbers(GER, vector_len))
            betti_GBA = np.array(betti_GBA) + np.array(betti_numbers(GBA, vector_len))
            betti_GWS = np.array(betti_GWS) + np.array(betti_numbers(GWS, vector_len))
        
        betti_GER = betti_GER / n_networks
        betti_GBA = betti_GBA / n_networks
        betti_GWS = betti_GWS / n_networks
        
        Y_GER.append(betti_GER)
        Y_GBA.append(betti_GBA)
        Y_GWS.append(betti_GWS)
            
    
    Y_GER = list(map(list, zip(*Y_GER)))
    Y_GBA = list(map(list, zip(*Y_GBA)))
    Y_GWS = list(map(list, zip(*Y_GWS)))
    
    plt.figure(figsize=(12,4))
    
    # Loop for para plotar os gráficos
    for i in range(vector_len):
        
        plt.subplot(1,3,1)
        plt.title('ER')
        plt.plot(X, Y_GER[i], label=f'Dim {i}')
        plt.legend()
    
        plt.subplot(1,3,2)
        plt.title('BA')
        plt.plot(X, Y_GBA[i], label=f'Dim {i}')
        plt.legend()
    
        plt.subplot(1,3,3)
        plt.title('WS')
        plt.plot(X, Y_GWS[i], label=f'Dim {i}')
        plt.legend()
        
    plt.show()
```

## Comparativo utilizando classificação de redes

Nesta seção são realizadas as computações das homologias simpliciais de algumas redes reais. Além disso, utilizamos métodos de classificação para tentar classificar essas redes utilizando as dimensões simpliciais e os números de Betti.

### Função de classificação de redes

A função a seguir, realiza a classificação de redes empregando o método dos k-vizinhos mais próximos em conjunto com três métodos de redução de dimensão: PCA, LDA e UMAP.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import umap

def classifier(y, X, X_test, k_neighbors, cl=['ER','WS','BA'], colors=['red', 'blue', 'green']):
    """
    Classificador k-vizinhos

    Args:
        y (array-like): Vetor de rótulos/targets.
        X (array-like): Matriz de características.
        X_test (array-like): Vetor de teste.
        n_neighbors (int): Quantidade de vizinhos a ser considerada.
        cl (list, optional): Lista de rótulos de classe. Defaults to ['ER','WS','BA'].
        colors (list, optional): Lista de cores para plotagem. Defaults to ['red', 'blue', 'green'].
    """
    y = np.array(y)
    # Padroniza os recursos removendo a média e dimensionando para a variância da unidade.
    scaler = StandardScaler().fit(X)
    X = scaler.transform(X)

    X_test = np.array(X_test)
    X_test = X_test.reshape(1,len(X_test)) 
    X_test = scaler.transform(X_test)

    # Métodos de redução de dimensão
    dim_reduction_methods = [    ("PCA", make_pipeline(PCA(n_components=2, random_state=0))),    ("LDA", make_pipeline(LinearDiscriminantAnalysis(n_components=2))),    ("UMAP", umap.UMAP())]

    knn = KNeighborsClassifier(n_neighbors=k_neighbors, metric='euclidean')

    plt.figure(figsize=(14, 4))

    # Plotagem para cada método de redução de dimensão
    for i, (name, model) in enumerate(dim_reduction_methods):
        plt.subplot(1, 3, i+1)

        # Ajusta o modelo do método
        model.fit(X, y)

        # Transforma os dados para 2 dimensões usando o modelo ajustado
        X_embedded = model.transform(X)
        X_test_embedded = model.transform(X_test)

        # Classificação usando KNN
        knn.fit(X_embedded, y)
        y_pred = knn.predict(X_test_embedded)

        # Encontra os k vizinhos mais próximos ao ponto central
        _, indices = knn.kneighbors(X_test_embedded, n_neighbors=k_neighbors)
        neighbors = X_embedded[indices][0]

        # Calcula o raio do círculo
        radius = np.max(np.linalg.norm(neighbors - X_test_embedded, axis=1))

        # Desenha um círculo em torno do ponto central com o raio calculado
        circle = plt.Circle(X_test_embedded[0], radius, color='black', fill=False, linestyle='dashed')
        plt.gca().add_artist(circle)

        print('Class:', cl[int(np.mean(y_pred))])

        for j, class_label in enumerate(cl):
            plt.scatter(
                X_embedded[y == j, 0],
                X_embedded[y == j, 1],
                c=colors[j],
                s=30,
                label=class_label
            )

        plt.scatter(
            X_test_embedded[:, 0],
            X_test_embedded[:, 1],
            c='black',
            s=100,
            marker="X",
            label="Test Case"
        )

        plt.legend()
        plt.title(name)
        plt.axis('equal')

    plt.show()
```

```python
import networkx as nx

def classifier_network(G):
    """Gera redes BA, ER e WS e realiza classificação utilizando medidas de rede e simpliciais.

    Args:
        G (networkx.Graph): Rede a ser classificada.
    """
    # Parâmetros de simulação
    n_nets = 30  # Quantidade de redes
    n_nodes = len(G)  # Quantidade de vértices
    av_degree = momment_of_degree_distribution(G, 1) # Grau médio
    k_neighbors = 6

    print(f"number of nodes={n_nodes} average degree={av_degree}")

    # Vetores para impressão de resultados
    X_measures = []
    X_betti_num = []
    X_simplex_dim = []
    y = []

    #ER networks
    p = av_degree/(n_nodes-1)
    for i in range(0,n_nets):
        GER = nx.gnp_random_graph(n_nodes, p, seed=None, directed=False)
        Gcc = sorted(nx.connected_components(GER), key=len, reverse=True)
        GER = GER.subgraph(Gcc[0])
        GER = nx.convert_node_labels_to_integers(GER, first_label=0)
        
        k1,k2,variance,av_cl,l,r = network_measures(GER)
        x = [k1,k2,variance,av_cl,l,r]
        X_measures.append(x)
        
        complex = get_simplicial_complex(GER)
        x = betti_numbers_of_complex(complex, vector_len=5)
        X_betti_num.append(x)
        
        x = simplexes_resume(complex, vector_len=11)
        X_simplex_dim.append(x)
        
        y.append(0)

    #WS networks
    k = int(av_degree)
    p = 0.05 #probability of rewiring
    for i in range(0,n_nets):
        GWS = nx.watts_strogatz_graph(n_nodes, k, p, seed=None)
        Gcc = sorted(nx.connected_components(GWS), key=len, reverse=True)
        GWS = GWS.subgraph(Gcc[0])
        GWS = nx.convert_node_labels_to_integers(GWS, first_label=0)
            
        k1,k2,variance,av_cl,l,r = network_measures(GWS)
        x = [k1,k2,variance,av_cl,l,r]
        X_measures.append(x)
        
        complex = get_simplicial_complex(GWS)
        x = betti_numbers_of_complex(complex, vector_len=5)
        X_betti_num.append(x)
        
        x = simplexes_resume(complex, vector_len=11)
        X_simplex_dim.append(x)
        
        y.append(1)

    # BA networks
    m = int(av_degree/2)
    for i in range(0,n_nets):
        GBA = nx.barabasi_albert_graph(n_nodes, m)    
        Gcc = sorted(nx.connected_components(GBA), key=len, reverse=True)
        GBA = GBA.subgraph(Gcc[0])
        GBA = nx.convert_node_labels_to_integers(GBA, first_label=0)
        
        k1,k2,variance,av_cl,l,r = network_measures(GBA)
        x = [k1,k2,variance,av_cl,l,r]
        X_measures.append(x)
        
        complex = get_simplicial_complex(GBA)
        x = betti_numbers_of_complex(complex, vector_len=5)
        X_betti_num.append(x)
        
        x = simplexes_resume(complex, vector_len=11)
        X_simplex_dim.append(x)
        
        y.append(2)
        
    print("network measures classification")
    k1,k2,variance,av_cl,l,r = network_measures(G)
    X_test = [k1, k2, variance, av_cl, l, r]
    classifier(y, X_measures, X_test, k_neighbors)

    complex = get_simplicial_complex(G)

    print("Betti numbers classification")
    X_test = betti_numbers_of_complex(complex, vector_len=5)
    classifier(y, X_betti_num, X_test, k_neighbors)

    print("Simplicial dimension classification")
    X_test = simplexes_resume(complex, vector_len=11)
    classifier(y, X_simplex_dim, X_test, k_neighbors)
```

```python
import gudhi as gd  
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

# Parâmetros dados
n_nodes = 100
av_degree = 9

# Parâmetros calculados
p = av_degree/(n_nodes-1)
m = int(av_degree/2)
k = int(av_degree)

# Erdos-Reni
GER = nx.erdos_renyi_graph(n_nodes, p, seed=None, directed=False)
GER.name = "Erdos-Renyi"
# Barabasi-Albert
GBA = nx.barabasi_albert_graph(n_nodes, m, seed=None)
GBA.name = "Barabasi-Albert"
# Small-world
GWS = nx.watts_strogatz_graph(n_nodes, k, p, seed=None)
GWS.name = "Watts-Strogatz"
    
plt.figure(figsize=(12, 4))

plt.subplot(1,3,1)
plt.title("GER")
pos = nx.spring_layout(GER, seed=42)
d = dict(GER.degree())
nx.draw(GER, pos, node_size=[v * 5 for v in d.values()], edgecolors='black', with_labels=False)
    
plt.subplot(1,3,2)
plt.title("GBA")
pos = nx.spring_layout(GBA, seed=42)
d = dict(GBA.degree())
nx.draw(GBA, pos, node_size=[v * 5 for v in d.values()], edgecolors='black', with_labels=False)
        
plt.subplot(1,3,3)
plt.title("GWS")
pos = nx.spring_layout(GWS, seed=42)
d = dict(GWS.degree())
nx.draw(GWS, pos, node_size=[v * 5 for v in d.values()], edgecolors='black', with_labels=False)
    
plt.show()

distanceMatrixs={}

for G in [GER,GBA,GWS]:
    distanceMatrixs[G.name]=pd.DataFrame(dict(nx.shortest_path_length(G))).sort_index()
        
distanceMatrixsMDS={}
n_components=4
for name in distanceMatrixs:
    try:
        embedding = MDS(n_components = n_components, dissimilarity = 'precomputed', normalized_stress='auto')
        distanceMatrixsMDS[name] = embedding.fit_transform(distanceMatrixs[name])
    except Exception as e:
        print(f"Error occurred during MDS embedding for {name}: {e}")

dim=5 #max dimension # HUGE impact on time complexity
simplexTrees={}
barCodes={}
for G in [GER,GBA,GWS]:
    alpha_complex = gd.AlphaComplex(
        points = distanceMatrixsMDS[G.name]
    )     

    Alpha_simplex_tree = alpha_complex.create_simplex_tree()
    simplexTrees[G.name]=Alpha_simplex_tree
    
    BarCode = Alpha_simplex_tree.persistence()
    barCodes[G.name]=BarCode

    plt.figure(figsize=(12,16),dpi=350)
    gd.plot_persistence_barcode(BarCode);    
    
    # Simplexes Resume
    simplexesLen={}
    for simplex,filtration in Alpha_simplex_tree.get_filtration():
        if(len(simplex)-1 not in simplexesLen):
            simplexesLen[len(simplex)-1]=1
        else:
            simplexesLen[len(simplex)-1]+=1    

    simplexesResume=''
    for item in simplexesLen.items():
        simplexesResume+=str(item[0])+': '+str(item[1])+' | '
    simplexesResume[0:-3]  
    
    # Barcode Resume
    
    dimensionsLen={}
    for dim in range(0,dim+1):
        dimensionsLen[dim]=len([bd for bd in BarCode if (bd[0]==dim)])

    dimensionResume=''
    for item in dimensionsLen.items():
        dimensionResume+=str(item[0])+': '+str(item[1])+' | '
    dimensionResume[0:-3]    
            
    # Plot
    
    plt.title(G.name+' n_components:'+str(n_components)+'\n Simplexes: '+simplexesResume[0:-3]+'\n Numbers of Bars: '+dimensionResume[0:-3],fontsize=9)
    plt.tight_layout()
    plt.show()
```
