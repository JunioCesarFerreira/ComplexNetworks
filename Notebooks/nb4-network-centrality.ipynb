Network Centrality

```python
from numpy  import *
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
np.random.seed(100)
```

```python
G= nx.read_edgelist("data/lesmis.txt", nodetype=int, data=(('weight',float),))
G = G.to_undirected()
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])
G = nx.convert_node_labels_to_integers(G, first_label=0)
N = len(G)
M = G.number_of_edges()
print('Number of nodes:', N)
print('Number of edges:', M)
plt.figure(figsize=(10,10))
pos=nx.spring_layout(G)
nx.draw(G, with_labels = True, pos = pos)
plt.show(True)
```

Random walk accessibility

```python
from scipy.linalg import expm

#G = nx.karate_club_graph()

#G = nx.Graph()
#G.add_edge('1','2'); G.add_edge('1','3'); G.add_edge('1','4'); G.add_edge('1','5'); G.add_edge('2','3'); 
#G.add_edge('4','5'); G.add_edge('2','6'); G.add_edge('3','7'); G.add_edge('4','8'); G.add_edge('5','9')
#G.add_edge('6','10'); G.add_edge('6','11'); G.add_edge('7','12'); G.add_edge('7','13'); G.add_edge('8','14')
#G.add_edge('8','15'); G.add_edge('9','16');G.add_edge('9','17')

N = 10
G = nx.star_graph(N)
N = len(G.nodes)

def acc(G):
    vk = dict(G.degree())
    vk = list(vk.values())
    A = nx.adjacency_matrix(G)
    P = np.zeros((N,N), dtype = 'float')
    for i in np.arange(0, N):
        for j in np.arange(0, N):
            if(vk[i] > 0):
                P[i,j] = A[i,j]/vk[i]
    P2 = expm(P)/np.exp(1)
    vacc = np.zeros(N, dtype = float)
    for i in np.arange(0, N):
        acc = 0
        for j in np.arange(0,N):
            if(P2[i,j] > 0):
                acc = acc + P2[i,j]*log(P2[i,j])
        acc = np.exp(-acc)
        vacc[i] = acc
    return vacc
print(acc(G))

plt.figure(figsize=(6,6))
pos=nx.spring_layout(G)
nx.draw(G, with_labels = True, pos = pos)
plt.show(True)
```

```python
# For the star graph
# Accessibility of the hub
x = np.cosh(1)/np.exp(1)
y = np.sinh(1)/np.exp(1)
acchub = exp(-x*np.log(x) -y*np.log(y/(N-1)))
print('Hub acc:', acchub)
# Accessibility of the leaves
x = np.sinh(1)/np.exp(1)
y = (cosh(1)-1)/(np.exp(1)*(N-1))
accleaves = exp(-x*np.log(x) - (N-2)*y*np.log(y) - (1/np.exp(1) + y)*np.log(1/np.exp(1) + y))
print('Leaves acc:', accleaves)
```

Localization

```python
N = 500 #Number of nodes
c = 4 #average degree
d = N/3 #degree of the main hub

def random_hub(N, c, d):
    p = c/((N-1)-1)
    G = nx.gnp_random_graph(N-1,p)
    G.add_node('hub')
    for i in G.nodes():
        if(np.random.uniform(0,1,1) < d/(len(G) - 1)):
            if(i != 'hub'):
                G.add_edge('hub',i,weight=1)
    return G

G = random_hub(N,c,d)
pos=nx.spring_layout(G)
pos['hub'] = [1.5,1.5]
k = G.subgraph('hub')  
plt.figure()
nx.draw_networkx(G, pos=pos, node_color = 'b', with_labels = False, node_size=50)
nx.draw_networkx(k, pos=pos, node_color = 'r', node_size=500)
plt.show(True)
```

```python
def IPR(v):
    ipr = 0
    z = 0
    for i in v:
        ipr = ipr + i**4
        z = z + i**2
    ipr = ipr/z**2
    return ipr

S = []
vd = []
T = c*(c+1)
for d in arange(1, 10*T,T/3):
    G = random_hub(N,c,d)
    aux = nx.eigenvector_centrality(G, max_iter = 1000)
    EC = []
    for i in G.nodes():
        EC.append(float(aux[i]))
    EC = np.array(EC)
    S.append(IPR(EC))
    vd.append(d)
S = np.array(S)
vd = np.array(vd)
plt.figure(figsize=(8,6))
plt.plot(vd,S,'ro')
plt.plot(vd,S,'b--')
plt.axvline(x=c*(c+1))
plt.xlabel("d", fontsize=20)
plt.ylabel("S", fontsize=20)
plt.title("Inverse Participation Ratio S", fontsize=20)
plt.grid(True)
plt.savefig('IPR.eps') #save the figure into a file
plt.show(True)
```
